```python
import pandas as pd

df = pd.read_csv('../data/pizza_types.csv', encoding='cp1250')
df
```

```python
df['_ing_list'] = df['ingredients'].str.split(',')

uniq = (df['_ing_list']
            .explode()
            .str.strip()
            .drop_duplicates()
            .reset_index(drop=True))

ingredients_df = pd.DataFrame({
        'ingredient_id': uniq.index + 1,
        'ingredient_name': uniq
})
```

```python
link = df[['pizza_type_id', '_ing_list']].explode('_ing_list')
link['ingredient_name'] = link['_ing_list'].str.strip()
link = link.merge(ingredients_df,
                      on='ingredient_name',
                      how='left')
pizza_ing_df = link[['pizza_type_id', 'ingredient_id']]

df = df.drop(columns=['ingredients', '_ing_list'])

pizza_ing_df
```

```python
ingredients_df.to_csv('../data/ingredients.csv', index=False)
pizza_ing_df.to_csv('../data/pizza_ingredients.csv', index=False)
df.to_csv('../data/pizzas_types_update.csv', index=False)
```
